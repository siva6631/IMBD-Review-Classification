```python
import re,nltk
```

```python
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
cv=CountVectorizer()
tf=TfidfVectorizer()
ps=PorterStemmer()
wl=WordNetLemmatizer()
```

```python
stop_words = stopwords.words('english')
print(stop_words[:15000]) 
```

```python
corpus=[]
for i in range(len(df)):
    rp=re.sub('[^a-zA-Z]',' ',df['review'].iloc[i])
    rp=rp.lower()
    rp=rp.split()
    rp=[wl.lemmatize(word) for word in rp if not word in set(stopwords.words('english'))-{'not','nor','no'
                                                                                          ,'never','didn',
                                                                                          "didn't",'aren',
                                                                                          "aren't",'can','couldn',"couldn't",'hadn',"hadn't","won't",
                                                                                         'should', 'shouldn', "shouldn't"}]
    rp=' '.join(rp)
    corpus.append(rp)
```

```python
corpus
```

```python
#vectorization
```

```python
cv=CountVectorizer(max_features=10000)
tf=TfidfVectorizer()
```

```python
x=cv.fit_transform(corpus).toarray()
x
```

```python
x.shape
```

```python
df['sentiment'].value_counts()
```

```python
df['sentiment']=df['sentiment'].replace({'positive':1,'negative':0})
```

```python
df['sentiment'].value_counts()
```

```python
y=df['sentiment']

```

```python
y
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import MultinomialNB
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=True)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
#saving model
```

```python
model=MultinomialNB()
```

```python
model.fit(x_train,y_train)
ypred_train=model.predict(x_train)
print('train:',accuracy_score(y_train,ypred_train))
print('cv:',cross_val_score(model,x_train,y_train,cv=5,scoring='accuracy').mean())
```

```python
ypred_test=model.predict(x_test)
print('test:',accuracy_score(y_test,ypred_test))
```
